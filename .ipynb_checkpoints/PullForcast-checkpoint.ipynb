```python
gh = github3.login(token=GITHUB_TOKEN)

f = open(FILENAME_ISSUES, 'w')
for issue in gh.search_issues('type:pr repo:SPM587SP19/SCM587SP19'):          # Find pulls from given Repo
            label_name=[]
            data={}
            current_issue = issue.as_json()
            current_issue = json.loads(current_issue)
            data['PullCount'] = 1
            data['created_at']= current_issue["created_at"][0:10]                 # Get created date of issue
            if current_issue["closed_at"] == None:
                data['closed_at']= current_issue["closed_at"]
            else:
                data['closed_at']= current_issue["closed_at"][0:10]               # Get closed date of issue
            data['State'] = current_issue["state"]                                # It gives state of issue like closed or open
            out=json.dumps(data)                                                  # save this all information to a JSON file
            f.write(out+ '\n')
f.close()
```

```python
import json
listOfPulls = [json.loads(line) for line in open('SPM587SP19_pull2.json')]
pull_df = DataFrame(listOfPulls)
pull_df.head()
```

```python
pull_df['created_at'] = pd.to_datetime(pull_df['created_at'], errors='coerce')

pull_df_created = pull_df.groupby(pull_df['created_at'].dt.date).sum().reset_index()
pull_df_created
```

```python
pull_df_created = pull_df_created.rename(index=str, columns={"created_at": "ds", "PullCount": "y"})
pull_df_created = pull_df_created[['ds', 'y']]
```

```python
ax = pull_df_created.set_index('ds').plot(figsize=(12, 8))
ax.set_ylabel('PR created')
ax.set_xlabel('Date')

plt.show()
```

```python
pullCreate = Prophet()
pullCreate.fit(pull_df_created)
```

```python
future = pullCreate.make_future_dataframe(periods=30)
future.tail()
```

```python
forecast = pullCreate.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = pullCreate.plot(forecast)
```

```python
fig1 = pullCreate.plot_components(forecast)
```

```python
pull_df['closed_at'] = pd.to_datetime(pull_df['closed_at'], errors='coerce')

pull_df_closed = pull_df.groupby(pull_df['closed_at'].dt.date).sum().reset_index()
pull_df_closed
```

```python
pull_df_closed = pull_df_closed.rename(index=str, columns={"closed_at": "ds", "PullCount": "y"})
pull_df_closed = pull_df_closed[['ds', 'y']]
pull_df_closed
```

```python
ax = pull_df_closed.set_index('ds').plot(figsize=(12, 8))
ax.set_ylabel('PR closed')
ax.set_xlabel('Date')

plt.show()
```

```python

pulls = Prophet()
pulls.fit(pull_df_closed)
future = pulls.make_future_dataframe(periods=30)
future.tail()
```

```python
forecast = pulls.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = pulls.plot(forecast)
```

```python
fig1 = pulls.plot_components(forecast)
```
